```python
import warnings
warnings.filterwarnings("ignore")


import numpy as np
import random

# # Set random seeds for reproducibility
seed = 42
np.random.seed(seed)
random.seed(seed)


# Task 1: Read the text from a file
# filepath = "text.txt"


# Concatenate text segments into a single string
file = open('./harry.txt','r')

final_text = file.read()
print(final_text)
```

```python
from transformers import pipeline

# Use a pipeline as a high-level helper
model_name = 'deepset/minilm-uncased-squad2'
deepset_pipeline = pipeline('question-answering',model=model_name,tokenizer=model_name)
```

```python
# Get a question from the user
query = "How did JK Rowlings got the idea of Harry Potter?"

result = deepset_pipeline(context=final_text,question=query)

# Display the answer
answer = result['answer']
# Perform document-question-answering
print(answer)
```
